```python
# Imports

import pandas as pd
import seaborn as sb
from matplotlib import pyplot
from adjustText import adjust_text
```

```python
# Variables
Directory = "data/"

filepath_experts = Directory + "GPS_Expert.sav"
filepath_parties = Directory + "GPS_Party.sav"
experts_df = pd.read_spss(filepath_experts)
parties_df = pd.read_spss(filepath_parties)
```

```python
def createMarkerFromCountry(country):
    if (country == "Canada") :
        return u'\U0001f341'
    if (country == "Australia") :
        return u'\2605'
    else:
        return u'\25CD'
    
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd            
```

```python
parties_df['Marker'] = parties_df['Country'].apply(createMarkerFromCountry)
canada_df = parties_df[parties_df['Country'] == 'Canada']
aus_df = parties_df[parties_df['Country'] == 'Australia']
nz_df = parties_df[parties_df['Country'] == 'New Zealand']
uk_df = parties_df[parties_df['Country'] == 'United States']
us_df = parties_df[parties_df['Country'] == 'United Kingdom']
commonwealth_df = pd.concat([canada_df, aus_df, nz_df, uk_df])
anglo_american_df = pd.concat([commonwealth_df, us_df])
```
